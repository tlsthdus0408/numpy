```python
fname=open("iris.csv")
```

```python
data1=fname.readline()# 하나씩 읽는 방법
print (data1)
```

```python
f = open("iris.csv")
data1 = f.readline()  # 하나씩 읽는 방법
print(data1)
data2 = []
for line1 in f:
    l = line1.strip().split(",")
    l[0] = float(l[0])
    l[1] = float(l[1])
    l[2] = float(l[2])
    l[3] = float(l[3])
    if l[4] == "Iris-setosa":    # 0
        l[4] = 0
    elif l[4] == "Iris-versicolor":   # 1
        l[4] = 1
    else:
        l[4] = 2
    data2.append(l)
f.close()
iris = np.array(data2)
```

```python
iris_labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
plt.plot(iris[:,0])
plt.title("SepalLength")
```

```python
plt.plot(iris[:,0])
plt.plot(iris[:,1])
plt.title("SepalLength&Speal Width")
plt.legend(["SepalLength","Speal Width"])
```

```python
plt.plot(iris[:,0])
plt.plot(iris[:,1])
plt.plot(iris[:,2])
plt.plot(iris[:,3])
plt.title("SepalLength&Speal Width")
plt.legend(["SepalLength","Speal Width, pw, es"])
```

```python
plt.scatter(iris[:,2], iris[:,3], c=iris[:,-1])
plt.colorbar()
plt.title('Scatter Plot')
plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
```

```python
plt.plot(iris[0,:-1], 'ro-') # setosa
plt.plot(iris[50,:-1], 'go-')
plt.plot(iris[100,:-1], 'bo-')
plt.legend(['setosa', 'versicolor', 'virginica'])
```

```python
plt.plot(iris[:50,:4].T, 'r-', alpha=0.1)
plt.plot(iris[50:100,:4].T, 'g-', alpha=0.1)
plt.plot(iris[100:,:4].T, 'b-', alpha=0.1)
plt.xticks(range(4),['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'])
```
